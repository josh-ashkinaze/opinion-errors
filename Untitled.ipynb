```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
```

```python
def generate_data(N, drift, lags= [0.5 for t in range(1, 2)], seed=1997):
    """
    Returns simulations from an AIMA model 
    
    Params
        N: Number of periods
        drift: A float for drift of time series
        lags: A list in form [1st lag beta, 2nd lag beta, ...]
        seed: Numpy random seed (if not none, will generate same data)
        
    Returns:
        A dataframe with column 'y' as value
        
    """
    np.random.seed(seed=seed)

    # Std normal noise 
    w = np.random.normal(loc=0, scale=1, size=N)

    # Init zero vector
    y = [drift] + [0]*(N-1)

    for t in range(1,N):

        # Add current noise
        y[t] += w[t]
   
        # Add past information
        for lag in range(1, len(lags)):
            # Lagged values
            y[t] += lags[lag]*y[t-lag]
            # Lagged errors 
            y[t] += lags[lag]*w[t-lag]

    
    df = pd.DataFrame([y]).T
    df.columns=['y']
    return df


lags= [0.1 for t in range(1, 11)]
risk = generate_data(N=1000, drift=1, lags=lags)

risk['y'].plot(figsize=(20, 10))
```

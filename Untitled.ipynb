```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
```

```python
def generate_data(N, drift, intercept=0, seed=None, lags= [0] + [1*(0.5**t) for t in range(1,20)]):
    """
    Returns simulations from an ARIMA model 
    
    Params
        N: Number of periods
        drift: A float for drft of time series
        intercept: Y intercept
        seed: Numpy random seed (if not none, will generate same data)
        lags: A list in form [1st lag beta, 2nd lag beta, ...]
        
    Returns:
        A dataframe with column 'y' as value
        
    """
    np.random.seed(seed=seed)

    # Std normal noise 
    w = np.random.standard_normal(size=N)

    # Init zero vector
    y = [intercept] + [0]*(N-1)

    
    for t in range(1,N):
        # Add drift 
        y[t] +=drift
        
        # Add current noise
        y[t] += w[t]
   
        # Add past information
        for lag in range(len(lags)):
            # Lagged values
            y[t] += lags[lag]*y[t-lag]
            # Lagged errors 
            y[t] += lags[lag]*w[t-lag]

    
    df = pd.DataFrame([y]).T
    df.columns=['y']
    return df

risk = generate_data(N=1000, intercept=1, drift=0.1, seed=None)


```

```python
class Agent:
    def __init__(self, uid, look_up, adjust, bias, truth):
        self.uid = uid
        self.look_up_threshold = look_up
        self.bias = bias
        self.adjust = adjust
        self.data=truth
        self.beliefs = [bias] + [0]*(len(truth)-1)
        
        
    def look_up(self, t):
        np.random.seed(seed=1997)
        if t > 0:
            if (self.data[t] - self.data[t-1])/(self.data[t-1]) > self.look_up_threshold:
                self.beliefs[t] = self.adjust*self.data[t] + (1-self.adjust)*self.beliefs[t]
            else:
                self.beliefs[t] = self.beliefs[t-1] + np.random.standard_normal(size=1)[0]
        
    def hello(self):
        print("Hello")
                  
    def get_beliefs(self): 
        for t in range(len(self.data)):
            self.look_up(t)
        df = pd.DataFrame([self.beliefs]).T
        df = df.fillna(method='ffill',  inplace=False)
        df['uid'] = self.uid
        df['truth'] = self.data
        df['t'] = df.index
        df.columns=['y', 'uid', 'truth', 't']
        return df

    


        
        
risk = generate_data(N=400, intercept=1, drift=0, seed=None)
a = Agent(uid=1, look_up=0, bias=2, adjust=0.3, truth=risk['y'].tolist())

beliefs = a.get_beliefs()

beliefs[['y', 'truth']].plot()
```

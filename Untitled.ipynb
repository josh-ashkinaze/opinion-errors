```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
```

```python
def generate_data(N, drift, lags= [0.5 for t in range(1, 2)], seed=None):
    """
    Returns simulations from an AIMA model 
    
    Params
        N: Number of periods
        drift: A float for drift of time series
        lags: A list in form [1st lag beta, 2nd lag beta, ...]
        seed: Numpy random seed (if not none, will generate same data)
        
    Returns:
        A dataframe with column 'y' as value
        
    """
    np.random.seed(seed=seed)

    # Array of normal noise 
    w = np.random.normal(loc=0, scale=1, size=N)

    # Array of zeros, starts with drift
    y = [drift] + [0]*(N-1)

    # Now we build the time series 
    for t in range(1,N):

        # Add drift
        y[t] += drift
        
        # Add past information
        for lag in range(1, len(lags)):
            if t >= lag:
#             # Add lagged values
                y[t] += lags[lag]*y[t-lag]
                # Add lagged errors 
                y[t] += lags[lag]*w[t-lag]

    
    df = pd.DataFrame([y]).T
    df.columns=['y']
    return df

n_lags = 20
lags = [0] + [0.5**x for x in range(1,n_lags+1)]
print(lags)
risk = generate_data(N=2000, drift=0.001, lags=lags)
risk['y'].plot(figsize=(20, 10))
display(risk['y'])
```

```python
class Agent:
    def __init__(self, uid, look_up, adjust,  bias, truth):
        self.uid = uid
        self.look_up_threshold = look_up
        self.bias = bias
        self.adjust = adjust
        self.data=truth
        self.beliefs = [bias] + [0]*(len(truth)-1)
        self.changes = self.get_changes()

        
        
    def look_up(self, t):
        if t > 0:
            np.random.seed(seed=None)
            
            # Beliefs follow a random walk 
            if abs(self.changes[t]) >= self.look_up_threshold:
                # If look up: Update beliefs to account for error
                self.beliefs[t] = self.data[t] 
            else:
                self.beliefs[t] = self.beliefs[t-1] + self.adjust*(self.data[t-1] - self.beliefs[t-1])

                

    def get_changes(self):
        truth_vals = pd.DataFrame(self.data)
        truth_vals['change'] = truth_vals[0].pct_change(periods=1)
        truth_vals = (truth_vals - truth_vals.mean())/truth_vals.std()
        return truth_vals['change'].tolist()
    
    def rmse(self, df):
        return ((df.er - df.er) ** 2).mean() ** .5


    
    @property
    def expectations(self): 
        for t in range(len(self.data)):
            self.look_up(t)
        df = pd.DataFrame([self.beliefs]).T
        df['uid'] = self.uid
        df['truth'] = self.data
        df['t'] = df.index
        df.columns=['belief', 'uid', 'truth', 't']
        df['er'] = (df['belief'] - df['truth'])
        rmse = (np.diff(df.er) ** 2).mean() ** .5
        
        return {"df":df, "rmse":rmse}

    



n_lags = 20
lags = [0] + [0.5**x for x in range(1,n_lags+1)]
risk = generate_data(N=2000, drift=0.1, lags=lags)
a = Agent(uid=1, look_up=1, adjust=0., bias=1,truth=risk['y'].tolist())
beliefs = a.expectations
```
